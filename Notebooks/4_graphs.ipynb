## Graphs
- Creating histograms of the RGB values of the most famous works that were correctly and incorrectly predicted.
- This will help understand how the color values might have influenced the model.

Link that helped me create a function to plot histograms for rgb values for the images
https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_histograms/py_histogram_begins/py_histogram_begins.html

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def histogram(path, number):
    img_path = path
    color = ('r', 'g', 'b')
    plt.figure(figsize=(10,10))
    for i,col in enumerate(color):
        histr = cv2.calcHist([cv2.imread(img_path)], [i], None, [256], [0,256])
        plt.plot(histr,color = col)
        plt.xlim([0,256])
    plt.show()
```

```python
renoir = './images2/81823.jpg'
renoir1 = '81823'

histogram(renoir, renoir1)
```

```python
renoir_luncheon_boating_party = './images2/54404.jpg'
renoir_luncheon_boating_party_num = '54404'

histogram(renoir_luncheon_boating_party, renoir_luncheon_boating_party_num)
```
